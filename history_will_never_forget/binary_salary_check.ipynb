### Idea: check that new data adds value by solving binary classification from the Kaggle kernel

```python
import pandas as pd

df = pd.read_csv('transformed_v2.csv', index_col=0)
```

```python
from pandas.api.types import CategoricalDtype

# transforming compensation into category type and ordening the values
categ = ['0-10,000', '10-20,000', '20-30,000', '30-40,000', '40-50,000',
         '50-60,000', '60-70,000', '70-80,000', '80-90,000', '90-100,000',
         '100-125,000', '125-150,000', '150-200,000', '200-250,000', '250-300,000',
         '300-400,000', '400-500,000', '500,000+']
cat_type = CategoricalDtype(categories=categ, ordered=True)
target = df.salary.astype(cat_type)
```

```python
map = {c: ('<100k' if i < 10 else '>100k') for i, c in enumerate(categ)}
```

```python
target_100k = target.replace(map)
```

```python
target_100k.value_counts(normalize=True)
```

```python
features = pd.get_dummies(df[[c for c in df.columns if c != 'salary']])
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```python
cross_val_score(LogisticRegression(random_state=42, solver='lbfgs', max_iter=1000),
                features, target_100k, cv=5).mean()
```

```python
cross_val_score(RandomForestClassifier(n_estimators=100), features, target_100k, cv=5).mean()
```

```python
target_50k = target.replace({c: ('<50k' if i < 4 else '>50k') for i, c in enumerate(categ)})
target_50k.value_counts(normalize=True)
```

```python
cross_val_score(LogisticRegression(random_state=42, solver='lbfgs', max_iter=1000),
                features, target_50k, cv=5).mean()
```

```python
cross_val_score(RandomForestClassifier(random_state=42, n_estimators=100),
                features, target_50k, cv=5).mean()
```

```python
rf = RandomForestClassifier(random_state=42, n_estimators=100)
rf.fit(features, target_50k)
```
